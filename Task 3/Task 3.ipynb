```python
df.drop_duplicates(subset=None,keep='first',inplace=True)
df.duplicated().sum()
```

```python
df['Postal Code'] = df['Postal Code'].astype('object')
```

```python
corr = df.corr()
sns.heatmap(corr,annot=True,cmap='Blues')
```

```python
df = df.drop(['Postal Code'],axis = 1)    #dropping postal code columns
```

```python
df.hist(bins=50 ,figsize=(20,15))
plt.show();
```

```python
sns.set(style="whitegrid")
plt.figure(2, figsize=(25,15))

sns.barplot(x='Sub-Category',y='Profit', data=df, palette='Spectral')

plt.suptitle('Consumption Patterns in the United States', fontsize=16)
```

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='State',data=df,palette='rocket_r',order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,8))
df['Region'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(20,8))
ax.scatter(df['Sales'],df['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['green','blue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
grouped=pd.DataFrame(df.groupby(['Ship Mode','Segment','Category','Sub-Category','State','Region'])['Quantity','Discount','Sales','Profit'].sum().reset_index())
grouped
```

```python
### sum,mean,min,max,count median,standard deviation,Variance of each states of Profit
df.groupby("State").Profit.agg(["sum","mean","min","max","count","median","std","var"])
```

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['green','blue'],figsize=(25,12))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

```python
t_states = df['State'].value_counts().nlargest(10)
t_states
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['blue','green'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['green','blue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```
